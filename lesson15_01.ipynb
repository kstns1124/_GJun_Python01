```python
#使用codespace 要安裝中文字型
#install 要加-y
#命令列:sudo apt-get update -y
#命令列:sudo apt-get install -y fonts-wqy-zenhei
#說明網址:https://askubuntu.com/questions/490829/how-can-i-install-chinese-fonts-on-kubuntu-14-04
#字型位置:/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc

import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
#mac: font=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=8)
#mac: font1=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=12)
#window: font=FontProperties(fname=r'c:\windows\fonts\simsun.ttc',size=8)

font1 = FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc',size=8)
font2 = FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc',size=16)
font3 = FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc',size=8)

student1 =  {'國文':78, '英文':80, '數學':92, '歷史':75, '探索':85}
student2 =  {'國文':59, '英文':95, '數學':59, '歷史':55, '探索':65}
figure = plt.figure(figsize=(8, 5), dpi=120, facecolor="#cccccc")
axes = figure.add_subplot(1,1,1)
axes.plot(student1.keys(),student1.values(), 'r--o', label='01')
axes.plot(student2.keys(),student2.values(), color='blue', marker='o', linestyle='dashed',
     linewidth=2, markersize=12, label='02')
axes.set_xticklabels(student1.keys(),fontproperties=font1)
axes.legend()
axes.set_title('成績',fontproperties=font2)
axes.set_xlabel("科目",fontproperties=font1)
axes.set_ylabel("分數",fontproperties=font1)

for x in axes.get_xticks():
    val1 = list(student1.values())[x]
    val2 = list(student2.values())[x]
    #print(x, ':', val)
    axes.annotate(text=str(val1), xy=(x-0.05, val1+1.0), color='red' ,fontproperties=font1
                  , arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.95', color='b'))
    axes.annotate(text=str(val2), xy=(x-0.05, val2+1.0), color='blue' ,fontproperties=font1
                  , arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.95', color='b'))
```

```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import random

#days = [num for num in range(1, 31)]
days = list(range(1, 31))
celsius = [round(random.uniform(0, 45), 1) for _ in range(1, 31)]

figure = plt.figure(figsize=(8, 5), dpi=120, facecolor="#cccccc")
axes = figure.add_subplot(1,1,1)
axes.plot(days, celsius, 'r', label='September', marker = 'H', markersize = 5)
axes.grid()
axes.set_title('Weathers',fontproperties=font2)
axes.set_xlabel("Day",fontproperties=font1)
axes.set_ylabel("Celsius",fontproperties=font1)
axes.legend()

axes.set_xticks(days)
plt.show()
```
