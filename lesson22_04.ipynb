```python
import numpy as np
import pandas as pd
```

```python
tips = pd.read_csv('csv_data/tips.csv')
```

```python
tips.columns =  ['總票價', '小費', '吸煙者', '日期', '時間', '大小']
```

```python
def roundVal(v):
    return round(number=v, ndigits=2)
tips['Ratio'] = tips['小費'] / tips['總票價']
tips['Ratio'] = tips['Ratio'].map(roundVal)
tips
```

```python
tips.sort_values(by='Ratio')[-6:]
```

```python
def top(df, n = 5, column = 'Ratio'):
    return df.sort_values(by=column)[-n :]
```

```python
top(tips, n=6, column='Ratio')
```

```python
tips.groupby(by=['吸煙者']).apply(top, n = 3)
```
