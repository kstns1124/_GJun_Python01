```python
# https://matplotlib.org/
import matplotlib.pyplot as plt

fig = plt.figure()

ax_1 = fig.add_subplot(1, 1, 1)
x = [0, 1, 2]
y = [6, 8, 10]
ax_1.set_title("Test Title")
ax_1.set_ylabel("Y values")
ax_1.set_xlabel("X values")
#ax_1.bar(x, y) #長條圖 
ax_1.plot(x, y) #線圖

plt.show()
```

```python
fig = plt.figure(figsize=(15, 3), dpi = 300)
ax_1 = fig.add_subplot(1, 3, 1)
ax_1.set_title("01")
ax_2 = fig.add_subplot(1, 3, 2)
ax_2.set_title("02")
ax_3 = fig.add_subplot(1, 3, 3)
ax_3.set_title("03")

plt.show()
```

```python
import random

fig = plt.figure(figsize=(5, 5), dpi = 100)
#students = [80, 59, 66, 81, 36]
students = [random.randint(0, 100) for x in range(5)]
ax = fig.add_subplot()
ax.plot(['chinese', 'math', 'english', 'history', 'create'], students)
plt.show()
```

```python
fig = plt.figure(figsize=(5, 5), dpi = 100)
students = {'chinese':59, 'math':69, 'english':49, 'history':87, 'create':58}
students2 = {'chinese':89, 'math':39, 'english':99, 'history':77, 'create':58}
ax = fig.add_subplot()
ax.plot(students.keys(), students.values())
ax.plot(students2.keys(), students2.values())
plt.show()
```
