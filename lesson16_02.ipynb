```python
import random
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import pandas as pd
```

```python
s01 = pd.Series(data = np.random.randint(low = 50, high = 100, size=5), index=['國文', '英文', '數學', '自然', '社會'])
s02 = pd.Series(data = [random.randint(50, 100) for _ in range(5)], index=['國文', '英文', '數學', '自然', '社會'])
s03 = pd.Series(data = [63, 69, 65, 75, 73], index=['國文', '英文', '數學', '自然', '社會'])
s04 = pd.Series(data = [55, 92, 72, 71, 72], index=['國文', '英文', '數學', '自然', '社會'])
s05 = pd.Series(data = [88, 91, 85, 83, 82], index=['國文', '英文', '數學', '自然', '社會'])
df = pd.DataFrame(index=['stuA', 'stuB', 'stuC', 'stuD', 'stuE']
                  , data=[s01, s02, s03, s04, s05]
                  , columns=['國文', '英文', '數學', '自然', '社會'])
df
```
