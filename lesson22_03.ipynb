```python
import numpy as np
import pandas as pd
```

```python
tips = pd.read_csv('csv_data/tips.csv')
tips
```

```python
tips.columns =  ['總票價', '小費', '吸煙者', '日期', '時間', '大小']
display(tips.info())
tips
```

```python
def roundVal(v):
    return round(number=v, ndigits=2)
tips['Ratio'] = tips['小費'] / tips['總票價']
tips['Ratio'] = tips['Ratio'].map(roundVal)
tips
```

```python
g01 = tips.groupby([tips['日期'], tips['時間'], tips['吸煙者']])
```

```python
d01 = dict(list(g01))
d01
```
