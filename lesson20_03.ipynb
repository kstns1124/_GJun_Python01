```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {'年廣告費投入':[12.5, 15.3, 23.2, 26.4, 33.5, 34.4, 39.4, 45.2, 55.4, 60.9],
            '月均銷售額':[21.2, 23.9, 32.9, 34.1, 42.5, 43.2, 49.0, 52.8, 59.4, 63.5]}

df = pd.DataFrame(data, index=[2011,2012,2013,2014,2015,2016,2017,2018,2019,2020])

fig = plt.figure(figsize=(8, 4));
axes = fig.add_subplot(1, 1, 1)
axes.plot(df.index, df['年廣告費投入'].values, 'ro--')
axes.plot(df.index, df['月均銷售額'].values, 'bx--')
plt.show()
```

```python
corr = df['年廣告費投入'].corr(df['月均銷售額'])
cov = df['月均銷售額'].corr(df['年廣告費投入'])
pd.DataFrame({'相關係數':[corr], '共變異數':[cov]})
```
