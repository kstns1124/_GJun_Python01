```python
frame1 = pd.DataFrame(np.random.randn(12).reshape((4, 3))
                  , index = [['a', 'a', 'b', 'b'], [1, 2, 1, 2]]
                  , columns = [['Taipei', 'Taipei', 'Tainan'], ['Green', 'blue', 'gray']])
display(frame1)
```

```python
frame1 = pd.DataFrame(np.arange(12).reshape((4,3)),
                      index=[['a','a','b','b'],[1, 2, 1, 2]],
                      columns=[['台北','台北','台中'],['Green','Red','Green']])
frame1
```

```python
frame1.index.names = ['key1', 'key2']
frame1
```

```python
frame1.columns.names = ['縣市','顏色']
frame1
```

```python
frame1['台北']
```

```python
frame1.swaplevel('key1', 'key2')
```

```python
frame1.sort_index(level=0, ascending=False)
```

```python
frame1.sort_index(level=1, ascending=False)
```

```python
#統計
frame1.groupby(level=0).sum()
```

```python
frame1.groupby(axis=1, level = '顏色').sum()
```

```python
frame1.T.groupby(level = '顏色').sum()
```

```python
frame1.T.groupby(level = '顏色').sum().T
```
