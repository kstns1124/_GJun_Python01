```python
import os
import json
import requests
import random
import csv

url = 'https://opendata.cwb.gov.tw/fileapi/v1/opendataapi/F-C0032-001?Authorization=rdec-key-123-45678-011121314&format=JSON'
```

```python
x = requests.get(url=url)
```

```python
# if Success, Save File
if (x.status_code == 200):
    print("下載成功!")
    #print(x.text)

    rootPath = os.path.abspath(path="./")
    fileName = os.path.join(rootPath, 'json_data', "20230903氣象_0.json")
    with open(file=fileName, mode='w', encoding='utf-8') as f:
        f.write(x.text)
```

```python
all_data = json.loads(x.text)
for i in all_data:
    print(i, all_data[i])
```

```python
print(all_data['cwbopendata']['sent'])
print(all_data['cwbopendata']['dataset']['datasetInfo']["datasetDescription"])
```

```python
locations = all_data['cwbopendata']['dataset']['location']
print(locations[4]['locationName'])
print(locations[4]['weatherElement'][0]['elementName'])
print(locations[4]['weatherElement'][0]['time'][0]['startTime'])
print(locations[4]['weatherElement'][0]['time'][0]['endTime'])
print(locations[4]['weatherElement'][0]['time'][0]['parameter'])
```

```python
weather_list = []
for item in locations:
    cityItem = {}
    cityItem['城市'] = item['locationName'];
    cityItem['開始時間'] = item['weatherElement'][0]['time'][0]['startTime']
    cityItem['結束時間'] = item['weatherElement'][0]['time'][0]['endTime']
    cityItem['最高溫度'] = item['weatherElement'][1]['time'][0]['parameter']['parameterName']
    cityItem['最低溫度'] = item['weatherElement'][2]['time'][0]['parameter']['parameterName']
    cityItem['感覺'] = item['weatherElement'][3]['time'][0]['parameter']['parameterName']
    weather_list.append(cityItem)
print(weather_list)
```

```python
with open("csv_data/Weather.csv", 'w', encoding='utf-8', newline='' )as f:
    writer = csv.DictWriter(f, fieldnames=list(weather_list[0].keys()));
    writer.writeheader()
    writer.writerows(weather_list)
print("OK!")
```
