```python
class Student:
    #初始化
    def __init__(self, name:str = "", chinese:int = 0, math:int = 0, eng:int = 0):
        self.__Name =  name
        self.Chinese = chinese
        self.Math = math
        self.English = eng
    #實體的方法
    def total(self)->int:
        return (self.Chinese +  self.Math + self.English)
     #實體的屬性
    @property
    def average(self)->int:
        avg = int(round(self.total()/3.0, ndigits= 0))
        return avg
    @property
    def Name(self) -> str:
        return self.__Name;    

    def __repr__(self):
        return f"姓名:{self.Name}\n國文:{self.Chinese}\n數學:{self.Math}\n英文:{self.English}"
```

```python
stu0 = Student()
stu1 = Student("DDD", 95, 88, 75)

stu2 = Student(name="OOO")
stu2.Chinese = 180
stu2.Math = 66
stu2.English = 100
```

```python
print(stu0.Name, stu0.Chinese,  stu0.Math, stu0.English, stu0.total(), stu0.average, sep = ", ")
print(stu1.Name, stu1.Chinese,  stu1.Math, stu1.English, stu1.total(), stu1.average, sep = ", ")
print(stu2.Name, stu2.Chinese,  stu2.Math, stu2.English, stu2.total(), stu2.average, sep = ", ")
```

```python
print(stu0.__repr__())
print(stu1.__repr__())
print(stu2)
```
