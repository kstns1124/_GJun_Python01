```python
import numpy as np
import pandas as pd
```

```python
f01 = pd.read_excel("excel/學生第一學期分數.xlsx")
```

```python
f02 = f01[['學號', '國文', '英文', '數學', '歷史', '地理']]
f02
```

```python
def description(s:pd.Series)->pd.Series:
    max = s.max()
    ind_max= s[s == max].index[0]
    min = s.min()
    ind_min= s[s == min].index[0]
    std = s.std();
    return pd.Series([ind_max, max,ind_min, min, std], index=['最高科目', '最高', '最低科目','最低', '標準差'])
```

```python
f03 = f02.set_index('學號')
f03
```

```python
f03.apply(description, axis= 1)
```

```python
f03.apply(description, axis= 0)
```
