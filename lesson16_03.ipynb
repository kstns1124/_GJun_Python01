```python
import random
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import pandas as pd
```

```python
scores = np.random.randint(low = 40, high=100, size = (50, 5))
```

```python
df0 = pd.DataFrame(data=scores
        , index=range(1, 51)
        , columns=['國文', '英文', '數學', '歷史', '地理'])
#df0
```

```python
print("sum:", df0.sum())
print("mean:", df0.mean())
print("max:", df0.max(), ", min:",df0.min())
print("median:", df0.median())
```
