```python
import random
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import pandas as pd
```

```python
x = np.arange(5)
y = np.array([5,2,9,3,8]) * 1000

font1 = FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc',size=12)

fig = plt.figure(figsize=(10,5))
axes1 = fig.add_subplot(1,1,1)
axes1.grid(axis='y')
axes1.bar(x, y)
axes1.set_xticks(x, ['apple', 'banana', 'orange', 'tomato', 'grava'])
axes1.legend()
# print(axes1.axis())
xmin,xmax, ymin,ymax = axes1.axis()
axes1.axis(xmin = xmin - (xmax-xmin)*0.05
    , xmax = xmax+ (xmax-xmin)*0.05
    , ymin = ymin
    , ymax = ymax + (ymax-ymin)*0.05)

for x in axes1.get_xticks():
    val1 = y [x]
    axes1.annotate(text=str(val1), xy=(x-0.15, val1+1.0*100), color='red' ,fontproperties=font1
                  #, arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.95', color='b')
                  )

plt.show()
```
